## Implement Core Chatbot Logic in Python

### Subtask:
Write the Python code to implement the chatbot's core logic, including a conversational loop, processing user input, and generating responses based on predefined rules for initial functionalities.


**Reasoning**:
The subtask requires implementing the core chatbot logic, including a conversational loop, processing user input, and generating responses based on predefined rules. I will provide a Python code block that defines the `run_chatbot()` function as per the instructions, incorporating an initial greeting, an infinite loop for user interaction, intent recognition using if/elif statements, static responses, and a fallback for unrecognized input.



````python
def run_chatbot():
    print("Hello! I'm your University Student Assistant. I'm here to help you with your academic life at the university. I can provide information on course schedules, academic resources, campus navigation, event information, study tips, university policies, and technical support. How can I assist you today?")

    while True:
        user_input = input("You: ").lower()

        if user_input in ['quit', 'bye']:
            print("Goodbye! Have a wonderful day!")
            break

        # Intent 1: Course Schedules
        elif "schedule" in user_input or "deadlines" in user_input:
            print("To get your personalized course schedule, I would need your student ID. Would you like to provide it, or are you looking for general academic dates?")

        # Intent 2: Study Tips
        elif "study tips" in user_input or "study advice" in user_input:
            print("I can offer some great study tips! Are you looking for general advice, or something specific like time management, exam preparation, or note-taking strategies?")

        # Intent 3: Academic Resources
        elif "academic resources" in user_input or "resources" in user_input:
            print("The university offers a variety of academic resources, including the writing center, tutoring services, and the library. Which one are you interested in learning more about?")

        # Fallback for unrecognized input
        else:
            print("I'm sorry, I don't understand that request. My apologies! I'm still learning. I can assist with topics related to: Course Schedules & Deadlines, Academic Resources, Campus Navigation, Event Information, Study Tips, University Policies, Technical Support. Please try rephrasing your question or choose from one of the topics above. You can also type 'menu' to see all my capabilities.")

# Call the chatbot function to start the interaction
run_chatbot()
````
